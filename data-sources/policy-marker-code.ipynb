```python
# build the timeline dataset
def build_policy_timeline_events():
    
    rows = [{"policy_id": "ADMIN_OBAMA",
            "policy_name": "Administration: Barack Obama",
            "category": "administration",
            "start_date": "2015-01-01",
            "end_date": "2017-01-19",
            "admin": "Obama",
            "notes": "Covers project window start through end of Obama term (inauguration transition).",
            "source": "https://www.whitehouse.gov/",
        },
            
        {
            "policy_id": "ADMIN_TRUMP_1",
            "policy_name": "Administration: Donald Trump (1st term)",
            "category": "administration",
            "start_date": "2017-01-20",
            "end_date": "2021-01-19",
            "admin": "Trump",
            "notes": "First Trump administration term.",
            "source": "https://www.whitehouse.gov/",
        },
            
        {
            "policy_id": "ADMIN_BIDEN",
            "policy_name": "Administration: Joe Biden",
            "category": "administration",
            "start_date": "2021-01-20",
            "end_date": "2025-01-19",
            "admin": "Biden",
            "notes": "Biden administration term (through Jan 19, 2025).",
            "source": "https://www.whitehouse.gov/",
        },
            
        {
            "policy_id": "ADMIN_TRUMP_2",
            "policy_name": "Administration: Donald Trump (2nd term)",
            "category": "administration",
            "start_date": "2025-01-20",
            "end_date": "2025-12-31",
            "admin": "Trump",
            "notes": "Second Trump administration period within the project window.",
            "source": "https://www.whitehouse.gov/",
        },
            
        {
            "policy_id": "NE_BORDER_2019",
            "policy_name": "National Emergency Concerning the Southern Border (Proclamation 9844)",
            "category": "national_emergency",
            "start_date": "2019-02-15",
            "end_date": "2021-02-11",
            "admin": "Trump/Biden",
            "notes": "Declared Feb 15, 2019; ended by Biden (ended Feb 11, 2021 per reporting).",
            "source": "https://time.com/5938570/border-emergency-military/",
        },
            
        {
            "policy_id": "PHE_COVID",
            "policy_name": "COVID-19 Public Health Emergency (HHS)",
            "category": "public_health_emergency",
            "start_date": "2020-01-31",
            "end_date": "2023-05-11",
            "admin": "Trump/Biden",
            "notes": "Federal COVID-19 PHE declared Jan 31, 2020; expired May 11, 2023.",
            "source": "https://www.hhs.gov/coronavirus/covid-19-public-health-emergency/index.html",
        },
            
        {
            "policy_id": "TITLE42_MAIN",
            "policy_name": "Title 42 border expulsions regime (CDC orders under 42 U.S.C. 265)",
            "category": "border_enforcement_public_health",
            "start_date": "2020-03-20",
            "end_date": "2023-05-11",
            "admin": "Trump/Biden",
            "notes": "Initial CDC order issued March 20, 2020; Title 42 ended May 11, 2023 alongside PHE expiration.",
            "source": "https://www.cdc.gov/port-health/legal-authorities/archived-orders.html",
        },
            
        {
            "policy_id": "TITLE42_ATTEMPT_END_2022",
            "policy_name": "CDC announced planned termination of Title 42 (later blocked)",
            "category": "border_enforcement_public_health",
            "start_date": "2022-04-01",
            "end_date": "2022-05-23",
            "admin": "Biden",
            "notes": "CDC released determination/order to terminate Title 42 on May 23, 2022; implementation was blocked by litigation.",
            "source": "https://www.dhs.gov/archive/news/2022/04/01/statement-secretary-mayorkas-cdcs-title-42-order-termination",
        },
            
        {
            "policy_id": "TITLE42_UAC_EXEMPT_2021",
            "policy_name": "Title 42 order updated to exempt unaccompanied children (UAC) (event marker)",
            "category": "border_enforcement_public_health",
            "start_date": "2021-08-02",
            "end_date": "2023-05-11",
            "admin": "Biden",
            "notes": "CDC replaced an October 2020 order with a new order continuing Title 42 while exempting unaccompanied children (UAC) per secondary summaries.",
            "source": "https://www.americanimmigrationcouncil.org/fact-sheet/guide-title-42-expulsions-border/",
        },
            
        {
            "policy_id": "CLP_RULE_2023",
            "policy_name": "Circumvention of Lawful Pathways (CLP) Final Rule (asylum processing)",
            "category": "asylum_processing",
            "start_date": "2023-05-11",
            "end_date": "2025-05-11",
            "admin": "Biden",
            "notes": "Final rule effective May 11, 2023; some summaries describe applicability through May 11, 2025.",
            "source": "https://www.federalregister.gov/documents/2023/05/16/2023-10146/circumvention-of-lawful-pathways",
        },
            
        {
            "policy_id": "MPP_1",
            "policy_name": "Migrant Protection Protocols (MPP 1.0 / 'Remain in Mexico')",
            "category": "asylum_processing",
            "start_date": "2019-01-01",
            "end_date": "2021-01-20",
            "admin": "Trump/Biden",
            "notes": "Went into effect in January 2019; suspended after Biden took office.",
            "source": "https://www.americanimmigrationcouncil.org/fact-sheet/migrant-protection-protocols/",
        },
            
        {
            "policy_id": "MPP_2",
            "policy_name": "Migrant Protection Protocols reinstatement (MPP 2.0) (returns to Mexico)",
            "category": "asylum_processing",
            "start_date": "2021-12-01",
            "end_date": "2022-10-31",
            "admin": "Biden",
            "notes": "Returned people to Mexico from Dec 2021 to Aug 2022; program formally ended again in Oct 2022.",
            "source": "https://www.americanimmigrationcouncil.org/fact-sheet/migrant-protection-protocols/",
        },
            
        {
            "policy_id": "MPP_3_2025",
            "policy_name": "Migrant Protection Protocols reimplemented (2025) (event marker)",
            "category": "asylum_processing",
            "start_date": "2025-01-21",
            "end_date": "2025-12-31",
            "admin": "Trump",
            "notes": "DHS announced MPP would be put back into effect in 2025 (details may evolve).",
            "source": "https://www.dhs.gov/news/2025/01/21/dhs-reinstates-migrant-protection-protocols",
        },
            
        {
            "policy_id": "MAYORKAS_PRIORITIES_2021",
            "policy_name": "DHS immigration enforcement priorities (Mayorkas guidelines effective Nov 29, 2021)",
            "category": "removal_priorities",
            "start_date": "2021-11-29",
            "end_date": "2025-12-31",
            "admin": "Biden",
            "notes": "Guidelines announced Sep 30, 2021; became effective Nov 29, 2021.",
            "source": "https://www.dhs.gov/archive/news/2021/09/30/secretary-mayorkas-announces-new-immigration-enforcement-priorities",
        },
    ]

    df = pd.DataFrame(rows)
    df["start_date"] = pd.to_datetime(df["start_date"])
    df["end_date"] = pd.to_datetime(df["end_date"])
    return df

# helper functions
def interval_overlaps_month(event_start, event_end, month_start):
    
    month_end = month_start + MonthEnd(1)
    return (event_start <= month_end) and (event_end >= month_start)

def active_in_month(timeline_df, policy_ids, month_start):
    
    subset = timeline_df[timeline_df["policy_id"].isin(policy_ids)]
    
    for _, row in subset.iterrows():
        if interval_overlaps_month(row["start_date"], row["end_date"], month_start):
            return 1
            
    return 0

def admin_on_month_start(timeline_df, month_start):
    
    admins = timeline_df[timeline_df["category"] == "administration"].copy()
    match = admins[
        (admins["start_date"] <= month_start) &
        (admins["end_date"] >= month_start)
    ]

    if match.empty:
        return None

    return match.iloc[0]["admin"]

# monthly marker dataset
def build_policy_markers_monthly(timeline_df, start = "2015-01-01", end = "2025-12-01"):
    
    months = pd.date_range(start=start, end=end, freq = "MS")
    rows = []
    
    for month_start in months:
        row = {
            "month": f"{month_start.month}/{month_start.day}/{month_start.year}",
            "admin": admin_on_month_start(timeline_df, month_start),
            "phe_covid": active_in_month(timeline_df, ["PHE_COVID"], month_start),
            "title42": active_in_month(timeline_df, ["TITLE42_MAIN"], month_start),
            "border_national_emergency": active_in_month(timeline_df, ["NE_BORDER_2019"], month_start),
            "mpp_active": active_in_month(timeline_df, ["MPP_1", "MPP_2", "MPP_3_2025"], month_start),
            "clp_rule_active": active_in_month(timeline_df, ["CLP_RULE_2023"], month_start),
            "mayorkas_priorities_active": active_in_month(timeline_df, ["MAYORKAS_PRIORITIES_2021"], month_start),
            "title42_uac_exempt": active_in_month(timeline_df, ["TITLE42_UAC_EXEMPT_2021"], month_start),
        }
        rows.append(row)

    return pd.DataFrame(rows)

# creating both datasets
policy_timeline_events_2015_2025 = build_policy_timeline_events()
policy_markers_monthly_2015_2025 = build_policy_markers_monthly(policy_timeline_events_2015_2025)

# save datasets
def save_outputs():
    
    policy_markers_monthly_2015_2025.to_csv("policy_markers_monthly_2015_2025.csv", index = False)
    policy_timeline_events_2015_2025.to_csv("policy_timeline_events_2015_2025.csv", index = False)

    print("Saved:")
    print(" - policy_markers_monthly_2015_2025.csv")
    print(" - policy_timeline_events_2015_2025.csv")

if __name__ == "__main__":
    print("Monthly shape :", policy_markers_monthly_2015_2025.shape)
    print("Timeline shape:", policy_timeline_events_2015_2025.shape)
    print("\nMonthly preview:")
    print(policy_markers_monthly_2015_2025.head())
    print("\nTimeline preview:")
    print(policy_timeline_events_2015_2025.head())
    save_outputs()
```
