```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
data = pd.read_excel("KHSM Repatriations fy25m11.xlsx", sheet_name=None)
```

```python
data.keys()
```

```python
annual_repats = data["Annual Repatriations"]
annual_repats.columns
```

```python
fig, ax = plt.subplots()
ax.plot(annual_repats["Fiscal\nYear"], annual_repats["Quantity"].apply(lambda x: x/1000000))
ax.set_title("Annual Repatriations")
ax.set_xlabel("Fiscal Year")
ax.set_ylabel("Quantity (Millions)")
ax.set_ylim((0,1.5))
```

```python
annual_repat_types = data["Annual Repatriation Type"]
annual_repat_types.columns
```

```python
figure = plt.figure(figsize=(20, 6))

repat_types = annual_repat_types["Repatriation Type"].unique()

num_plots = len(repat_types)

formatter = matplotlib.ticker.StrMethodFormatter("{x:.0f}")

for i, repat_type in enumerate(repat_types):
    axes = figure.add_subplot(1, num_plots, i+1)
    axes.bar(annual_repat_types[annual_repat_types["Repatriation Type"] == repat_type]["Fiscal\nYear"], annual_repat_types[annual_repat_types["Repatriation Type"] == repat_type]["Quantity"].apply(lambda x: x/1000))
    axes.set_title(f"{repat_type}")
    axes.set_ylim((0,1200))
    axes.set_xlim((2009,2026))
    plt.gca().xaxis.set_major_formatter(formatter)

figure.suptitle("Annual Repatriations by Type")
figure.supylabel("Quantity (Thousands)")
figure.supxlabel("Fiscal Year")
plt.show()
plt.close()
```

```python
encounters2023 = pd.read_csv("nationwide-encounters-fy20-fy23-state.csv")
encounters2124 = pd.read_csv("nationwide-encounters-fy21-fy24-state.csv")
encounters2526 = pd.read_csv("nationwide-encounters-fy23-fy26-jan-state.csv")
encounters2526 = encounters2526[encounters2526["Fiscal Year"].apply(lambda x: ("2023" not in x) & ("2024" not in x))]
encounters2526["Fiscal Year"] = encounters2526["Fiscal Year"].apply(lambda x: 2025 if ("2025" in x) else 2026)
```

```python
encounters = pd.concat([encounters2023, encounters2124,  encounters2526], axis=0)
encounters.columns
```

```python
encounters.head()
```

```python
yearly_encounters = encounters.groupby("Fiscal Year")["Encounter Count"].sum()
```

```python
fig, ax = plt.subplots()
ax.plot(yearly_encounters.keys(), yearly_encounters.values/1000000)
ax.set_title("Annual Encounters")
ax.set_xlabel("Fiscal Year")
ax.set_ylabel("Quantity (Millions)")
```

```python
figure = plt.figure(figsize=(20, 6))

authorities = encounters["Title of Authority"].unique()

for i, authority in enumerate(authorities):
    axes = figure.add_subplot(1, len(authorities), i+1)
    authority_encounters = encounters[encounters["Title of Authority"] == authority].groupby("Fiscal Year")["Encounter Count"].sum()
    axes.bar(authority_encounters.keys(), authority_encounters.values/1000000)
    axes.set_title(f"{authority}")
    axes.set_ylim((0,6))
    axes.set_xticks(range(2020, 2027))
    plt.gca().xaxis.set_major_formatter(formatter)

figure.suptitle("Annual Encounters by Title of Authority")
figure.supylabel("Quantity (Millions)")
figure.supxlabel("Fiscal Year")
```

```python
figure = plt.figure(figsize=(20, 6))

borders = encounters["Land Border Region"].unique()

for i, border in enumerate(borders):
    axes = figure.add_subplot(1, len(borders), i+1)
    border_encounters = encounters[encounters["Land Border Region"] == border].groupby("Fiscal Year")["Encounter Count"].sum()
    axes.bar(border_encounters.keys(), border_encounters.values/1000000)
    axes.set_title(f"{border}")
    axes.set_ylim((0,6))
    axes.set_xticks(range(2020, 2027))
    plt.gca().xaxis.set_major_formatter(formatter)

figure.suptitle("Annual Encounters by Land Border Region")
figure.supylabel("Quantity (Millions)")
figure.supxlabel("Fiscal Year")
```

```python
figure = plt.figure(figsize=(20, 12))

years = encounters["Fiscal Year"].unique()

for i, year in enumerate(years):
    axes = figure.add_subplot(2, 4, i+1)
    citizenships = encounters[encounters["Fiscal Year"] == year].groupby("Citizenship")["Encounter Count"].sum().sort_values(ascending=False)
    axes.bar(citizenships.keys()[:5], citizenships.values[:5]/1000000)
    axes.set_title(f"{year}")
    axes.set_ylim((0,1.75))
    axes.set_xticks(range(0,5))
    axes.set_xticklabels(citizenships.keys()[:5], rotation=15)

figure.suptitle("Most Frequent Citizenship of Encounters by Year")
figure.supylabel("Quantity (Millions)")
figure.supxlabel("Citizenship")
```

```python
figure = plt.figure(figsize=(20, 12))

years = encounters["Fiscal Year"].unique()

for i, year in enumerate(years):
    axes = figure.add_subplot(2, 4, i+1)
    citizenships = encounters[encounters["Fiscal Year"] == year].groupby("Demographic")["Encounter Count"].sum().sort_values(ascending=False)
    axes.bar(citizenships.keys(), citizenships.values/1000000)
    axes.set_title(f"{year}")
    axes.set_ylim((0,4.5))
    axes.set_xticks(range(0,4))
    axes.set_xticklabels(citizenships.keys(), rotation=10)

figure.suptitle("Demographics of Encounters by Year")
figure.supylabel("Quantity (Millions)")
figure.supxlabel("Demographic")
```

```python
figure = plt.figure(figsize=(20, 12))

years = encounters["Fiscal Year"].unique()

for i, year in enumerate(years):
    axes = figure.add_subplot(2, 4, i+1)
    states = encounters[encounters["Fiscal Year"] == year].groupby("State")["Encounter Count"].sum().sort_values(ascending=False)
    axes.bar(states.keys()[:5], states.values[:5]/1000000)
    axes.set_title(f"{year}")
    axes.set_ylim((0,3))
    axes.set_xticks(range(0,5))
    axes.set_xticklabels(states.keys()[:5], rotation=15)

figure.suptitle("Most Frequent State of Encounters by Year")
figure.supxlabel("State")
figure.supylabel("Quantity (Millions)")
```
